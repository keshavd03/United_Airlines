### Satisfaction vs flight originate

```python
df_satisfaction.columns
```

```python
df_temp = df_satisfaction[['origin_station_code', 'score']]
df_scores = df_temp.groupby('origin_station_code').mean().reset_index()
```

```python
df_counts = df_temp.groupby('origin_station_code').count().reset_index().sort_values('score', ascending=False).iloc[:10]
```

```python
df_counts = df_counts.reset_index().drop(columns = ['index']).rename(columns = {'score': 'count'})
df_counts
```

```python
df_scores
```

```python
df_temp = df_scores.merge(df_counts).sort_values('count', ascending=False)
df_temp
```
