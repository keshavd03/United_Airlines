```python
import pandas as pd
import numpy as np
import seaborn as sns 
```

```python
inventory= pd.read_csv('Inflight Service_Inventory data2.csv')
preorder= pd.read_csv('Inflight Service_Pre order data2.csv')
customer= pd.read_csv('Survey data _Customer comments2.csv')
satisfaction= pd.read_csv('Survey data_Inflight Satisfaction Score2.csv')
```

```python
satisfaction.columns
```

```python
inventory.rename(columns = {'departure_station_code':'origin_station_code', 'arrival_station_code':'destination_station_code', 'scheduled_departure_dtl':'scheduled_departure_date'}, inplace = True)
inventory.columns
```

```python
preorder.rename(columns = {'departure_station_code':'origin_station_code', 'arrival_station_code':'destination_station_code', 'scheduled_departure_dtl':'scheduled_departure_date'}, inplace = True)
preorder.columns
```

```python
df_merge= pd.merge(satisfaction, inventory, on = ['flight_number', 'origin_station_code', 'destination_station_code'])
df_merge
```

```python
df_merge= pd.merge(df_merge,preorder, on = ['flight_number', 'origin_station_code', 'destination_station_code', 'record_locator'])
df_merge
```
